```python
import pandas as pd # Importa a biblioteca pandas para manipulação de dados
import plotly.express as px # Importa a biblioteca plotly express para visualização de dados
```

```python
# Lê o arquivo Excel e armazena os dados da aba "Total_de_acoes" em um DataFrame chamado df_Total_de_acoes
df_Total_de_acoes = pd.read_excel("/content/ArquivoInicial-Aula2.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes # Exibe o DataFrame df_Total_de_acoes
```

```python
# Lê o arquivo Excel e armazena os dados da aba "Principal" em um DataFrame chamado df_Principal
df_Principal = pd.read_excel("/content/ArquivoInicial-Aula2.xlsx", sheet_name="Principal")
df_Principal  #Exibe o DataFrame df_Principal
```

```python
# Lê o arquivo Excel e armazena os dados da aba "Ticker" em um DataFrame chamado df_Ticker
df_Ticker = pd.read_excel("/content/ArquivoInicial-Aula2.xlsx", sheet_name="Ticker")
df_Ticker #Exibe o DataFrame df_Ticker
```

```python
# Lê o arquivo Excel e armazena os dados da aba "ChatGPT" em um DataFrame chamado df_ChatGPT
df_ChatGPT = pd.read_excel("/content/ArquivoInicial-Aula2.xlsx", sheet_name="ChatGPT")
df_ChatGPT #Exibe o DataFrame df_ChatGPT
```

```python
# Seleciona apenas as colunas 'Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)' do DataFrame df_Principal e cria uma cópia
df_Principal = df_Principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_Principal # Exibe o DataFrame df_Principal com as colunas selecionadas
```

```python
# Renomeia as colunas 'Último (R$)' para 'valor final' e 'Var. Dia (%)' para 'var_dia_pct' no DataFrame df_Principal e cria uma cópia
df_Principal = df_Principal.rename(columns={'Último (R$)':'valor final','Var. Dia (%)':'var_dia_pct'}).copy()
df_Principal # Exibe o DataFrame df_Principal com as colunas renomeadas
```

```python
# Calcula a variação percentual do dia ('var_dia_pct') dividida por 100 e armazena o resultado na coluna 'Var_pct'
df_Principal['Var_pct'] = df_Principal['var_dia_pct']/100
df_Principal # Exibe o DataFrame df_Principal com a nova coluna 'Var_pct'
```

```python
# Calcula o valor inicial utilizando a fórmula: Valor Inicial = Valor Final / (Variação Percentual + 1)
# Isso é uma forma de calcular o valor inicial quando você conhece a variação percentual e o valor final
df_Principal['Valor inicial'] = df_Principal['valor final']/(df_Principal['Var_pct'] +1)
df_Principal # Exibe o DataFrame df_Principal com as novas colunas 'Var_pct' e 'Valor inicial'
```

```python
# Mescla os DataFrames df_Principal e df_Total_de_acoes com base nas colunas 'Ativo' e 'Código'
# A mesclagem é feita da esquerda (df_Principal) para a direita (df_Total_de_acoes)
df_Principal = df_Principal.merge(df_Total_de_acoes, left_on='Ativo'	,right_on= 'Código', how='left' )
df_Principal
```

```python
# Remove a coluna 'Código' do DataFrame df_Principal
df_Principal = df_Principal.drop(columns=['Código'])
df_Principal # Exibe o DataFrame df_Principal após a remoção da coluna 'Código'
```

```python
# Calcula a variação em reais ('Variação r$') multiplicando a diferença entre o 'valor final' e o 'Valor inicial' pelo 'Qtde. Teórica'
df_Principal['Variação r$'] = (df_Principal['valor final'] - df_Principal['Valor inicial']) * df_Principal['Qtde. Teórica']
df_Principal # Exibe o DataFrame df_Principal com a nova coluna 'Variação r$'
```

```python
# Altera o tipo de dados da coluna 'Qtde. Teórica' para inteiro
df_Principal['Qtde. Teórica']=df_Principal['Qtde. Teórica'].astype(int)
df_Principal
```

```python
# Renomeia a coluna 'Qtde. Teórica' para 'qtd_teorica' no DataFrame df_Principal e cria uma cópia
df_Principal = df_Principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_Principal # Exibe o DataFrame df_Principal após a renomeação da coluna
```

```python
# Cria uma nova coluna 'resultado' com base na variação em reais ('Variação r$')
# Se a variação for maior que zero, o resultado será 'Subiu'
# Se a variação for menor que zero, o resultado será 'Desceu'
# Se a variação for igual a zero, o resultado será 'Estável'
df_Principal['resultado'] = df_Principal['Variação r$'].apply(lambda x: 'Subiu' if x>0 else('Desceu' if x<0 else 'Estável'))
df_Principal
```

```python
# Mescla os DataFrames df_Principal e df_Ticker com base nas colunas 'Ativo' e 'Ticker'
df_Principal = df_Principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_Principal = df_Principal.drop(columns=['Ticker']) # Remove a coluna 'Ticker' do DataFrame df_Principal
df_Principal
```

```python
 #Mescla os DataFrames df_Principal e df_ChatGPT com base nas colunas 'Nome' e 'Empresa'
df_Principal = df_Principal.merge(df_ChatGPT, left_on='Nome', right_on='Empresa', how='left')
df_Principal = df_Principal.drop(columns=['Empresa']) # Remove a coluna 'Empresa' do DataFrame df_Principal
df_Principal
```

```python
# Renomeia a coluna 'Idade (em anos)' para 'Idade' no DataFrame df_Principal e cria uma cópia
df_Principal = df_Principal.rename(columns={'Idade (em anos)':'Idade'}).copy()
df_Principal
```

```python

# Cria uma nova coluna 'Cat_idade' com base na idade ('Idade')
# Se a idade for maior que 100 anos, atribui 'Maior que 100 anos'
# Se a idade for menor que 50 anos, atribui 'Menos que 50'
# Caso contrário, atribui 'Entre 50 e 100'
df_Principal['Cat_idade'] = df_Principal['Idade'].apply(lambda x: "Maior que 100 anos" if x > 100 else ("Menos que 50" if x < 50 else "Entre 50 e 100"))
df_Principal
```

```python
# Calculando o maior valor
maior = df_Principal['Variação r$'].max()

# Calculando o menor valor
menor = df_Principal['Variação r$'].min()

# Calculando a média
media = df_Principal['Variação r$'].mean()

# Calculando a média de quem subiu
media_subiu = df_Principal[df_Principal['resultado'] == 'Subiu']['Variação r$'].mean()

# Calculando a média de quem desceu
media_desceu = df_Principal[df_Principal['resultado'] == 'Desceu']['Variação r$'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
# Filtra o DataFrame df_Principal para incluir apenas as linhas onde o valor da coluna 'resultado' é 'Subiu'
df_Principal_subiu = df_Principal[df_Principal['resultado'] == 'Subiu']
df_Principal_subiu
```

```python
# Calcula a soma da variação em reais ('Variação r$') agrupada pelo segmento ('Segmento')
# e armazena o resultado em um novo DataFrame df_analise_segmento
df_analise_segmento = df_Principal_subiu.groupby('Segmento')['Variação r$'].sum().reset_index()
df_analise_segmento
```

```python
# Calcula a soma da variação em reais ('Variação r$') agrupada pelo resultado ('resultado')
# e armazena o resultado em um novo DataFrame df_analise_saldo
df_analise_saldo = df_Principal.groupby('resultado')['Variação r$'].sum().reset_index()
df_analise_saldo
```

```python
# Cria um gráfico de barras utilizando o DataFrame df_analise_saldo
# O eixo x é definido como 'resultado', o eixo y como 'Variação r$' e o texto como 'Variação r$'
# O título do gráfico é definido como 'Variação Reais por Resultado'
fig = px.bar(df_analise_saldo, x='resultado', y='Variação r$', text='Variação r$', title='Variação Reais por Resultado')
fig.show() # Exibe o gráfico
```
