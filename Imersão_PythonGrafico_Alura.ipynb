```python
# Importa a biblioteca pandas e a renomeia como pd
import pandas as pd

# Importa a sub-biblioteca pyplot da biblioteca matplotlib e a renomeia como plt
import matplotlib.pyplot as plt

# Importa o módulo dates da biblioteca matplotlib e a renomeia como mdates
import matplotlib.dates as mdates

# Importa a biblioteca mplfinance e a renomeia como mpf
import mplfinance as mpf

# Importa a biblioteca yfinance e a renomeia como yf
import yfinance as yf

# Importa o módulo graph_objects da biblioteca plotly e a renomeia como go
import plotly.graph_objects as go

# Importa a função make_subplots da sub-biblioteca plotly.subplots e a renomeia como make_subplots
from plotly.subplots import make_subplots
```

```python
# Utiliza a função download do pacote yfinance para baixar dados históricos de preços das ações da PETR4.SA (Petrobras)
# O argumento start especifica a data de início dos dados (1º de janeiro de 2023)
# O argumento end especifica a data de término dos dados (31 de dezembro de 2023)
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
```

```python
# Renomeia as colunas do DataFrame de dados baixados
dados.columns = ['Abertura', 'Máximo', 'Mínimo', 'Fechamento', 'Fechamento Ajustado', 'Volume']
```

```python
# Renomeia o índice do DataFrame para 'Data'
dados = dados.rename_axis('Data')
```

```python
# Plota um gráfico de linha dos preços de fechamento ao longo do tempo
dados['Fechamento'].plot(figsize=(10, 6))

# Adiciona um título ao gráfico
plt.title('Variação do preço por data', fontsize=16)

# Adiciona uma legenda ao gráfico
plt.legend(['Fechamento'])
```

```python
# Cria uma cópia das primeiras 60 linhas do DataFrame dados
df = dados.head(60).copy()

# Adiciona uma coluna 'Data' ao DataFrame df e preenche com o índice (que são datas)
df['Data'] = df.index

# Converte as datas para o formato numérico aceito pelo Matplotlib
# Isso é necessário para que o Matplotlib possa plotar as datas corretamente no gráfico
df['Data'] = df['Data'].apply(mdates.date2num)

# Exibe o DataFrame df
df
```

```python
# Cria uma figura e eixo para o gráfico
fig, ax = plt.subplots(figsize=(15, 8))

# Define a largura dos candles no gráfico
width = 0.7

# Loop sobre os dados para plotar cada candle
for i in range(len(df)):
    # Determina a cor do candle com base na variação do preço (verde para valorização, vermelho para desvalorização)
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
        color = 'green'
    else:
        color = 'red'

    # Desenha a linha vertical do candle (mecha)
    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
            [df['Mínimo'].iloc[i], df['Máximo'].iloc[i]],
            color=color,
            linewidth=1)

    # Desenha o corpo do candle
    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                               width,
                               abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                               facecolor=color))

# Calcula e plota as médias móveis de 7 e 14 dias
df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()
ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 Dias')  # Média de 14 dias

# Adiciona legendas para as médias móveis
ax.legend()

# Formata o eixo x para mostrar as datas corretamente
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adiciona título e rótulos dos eixos x e y
plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

# Adiciona uma grade para facilitar a visualização dos valores
plt.grid(True)

# Exibe o gráfico
plt.show()
```

```python
# Criando subplots
'''
Primeiro, criamos uma figura que conterá nossos gráficos usando make_subplots.
Isso nos permite ter múltiplos gráficos em uma única visualização.
Aqui, teremos dois subplots: um para o gráfico de candlestick e outro para o volume de transações.
'''
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

# Adicionando o gráfico de candlestick
'''
No gráfico de candlestick, cada candle representa um dia de negociação,
mostrando o preço de abertura, fechamento, máximo e mínimo. Vamos adicionar este gráfico à nossa figura.
'''
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['Abertura'],
                             high=df['Máximo'],
                             low=df['Mínimo'],
                             close=df['Fechamento'],
                             name='Candlestick'),
                             row=1, col=1)

# Adicionando as médias móveis
# Adicionamos também médias móveis ao mesmo subplot para análise de tendências
fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA7'],
                         mode='lines',
                         name='MA7 - Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA14'],
                         mode='lines',
                         name='MA14 - Média Móvel 14 Dias'),
                         row=1, col=1)

# Adicionando o gráfico de barras para o volume
# Em seguida, criamos um gráfico de barras para o volume de transações, que nos dá uma ideia da atividade de negociação naquele dia
fig.add_trace(go.Bar(x=df.index,
                     y=df['Volume'],
                     name='Volume'),
                     row=2, col=1)

# Atualizando layout
# Finalmente, configuramos o layout da figura, ajustando títulos, formatos de eixo e outras configurações para tornar o gráfico claro e legível.
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False,  # Desativa o range slider
                  width=1100, height=600)

# Mostrando o gráfico
fig.show()
```
